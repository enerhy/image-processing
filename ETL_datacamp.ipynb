```python
from skimage import data, color
import matplotlib.pyplot as plt
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
coffee_image = data.coffee()
coins_image = data.coins()
rocket = data.rocket()
```

```python
show_image(coffee_image)
```

```python
print(coffee_image.shape)
print(coins_image.shape)
```

```python
gray_scaled_rocket = color.rgb2gray(rocket)
show_image(gray_scaled_rocket)
```

### Thresholding

#### Global Thresholding

```python
from skimage.filters import try_all_threshold

image_gray = color.rgb2gray(coffee_image)

# Obtain all the resulting 
imagesfig, ax = try_all_threshold(image_gray, verbose=False)

# Showing resulting plots
plt.show()
```

```python
image_gray = color.rgb2gray(coffee_image)
thresh = 20
binary = image_gray > thresh
inverse = image_gray <= thresh
plt.imshow(binary)
show_image(binary)
show_image(inverse)
```

```python
plt.hist(image_gray.ravel(), bins=255)
plt.show()
```

```python
from skimage.filters import threshold_otsu
# Make the image grayscale using rgb2gray
image_gray = color.rgb2gray(coffee_image)

# Obtain the optimal threshold value with otsu
thresh = threshold_otsu(image_gray)

# Apply thresholding to the image
binary = image_gray > thresh

# Show the image
show_image(binary, 'Binary image')
```

#### Local thresholding

```python
"""
To work we need not a binariza 2D gray image
"""

# Import the local threshold function
from skimage.filters import threshold_local

# Make the image grayscale using rgb2gray
image_gray = color.rgb2gray(coffee_image)

# sliced image
image_red = coffee_image[:,:, 0]

# Set the block size to 35
block_size = 35

# Obtain the optimal local image_red
local_thresh = threshold_local(image_red, block_size, offset=10)

# Obtain the binary image by applying local thresholding
binary_local = image_red > local_thresh

# Show the binary image
show_image(binary_local, 'Local thresholding')
```

```python
from PIL import Image

path = "data/tools_image.jpg"
tools_image = Image.open(path)
tools_image_array = np.asarray(tools_image)
print(type(tools_image))
print(type(tools_image_array))
```

```python
show_image(tools_image_array)
```

```python
gray_tools = color.rgb2gray(tools_image_array)
plt.hist(gray_tools.ravel(), bins=255)
plt.show()
```

```python
thresh_manual = 0.38

binary_image_manual = gray_tools > thresh_manual
show_image(binary_image_manual, 'Manual Binarized image')

thresh = threshold_otsu(gray_tools)
binary_image_otsu = gray_tools > thresh
show_image(binary_image_otsu, 'Automatic Binarized image')

```
